```python
def simple_calc():
    a=float(input("Enter the first number: "))
    b=float(input("Enter the second number: "))
    print("% means remainder")
    print("** means exponents")
    operation=input("Enter the operation (+, -, *, /,%,**): ")
    match operation:
        case "+":
            print("The sum of ",a," and ",b," is ",a+b)
        case "-":
            dif=input("a-b or b-a: ")
            if dif == "a-b":
                print("The difference of ",a," and ",b," is ",a-b)
            elif dif == "b-a":
                print("The difference of ",b," and ",a," is ",b-a)
            else:
                print("Invalid input")
        case "*":
            print("The product of ",a," and ",b," is ",a*b)
        case "/":
            div=input("a/b or b/a: ")
            if div == "a/b":
                if b != 0:
                    print("The division of ",a," by ",b," is ",a/b)
                else:
                    print("Error: Division by zero is not allowed.")
            elif div == "b/a":
                if a != 0:
                    print("The division of ",b," by ",a," is ",b/a)
                else:
                    print("Error: Division by zero is not allowed.")
            else:
                print("Invalid input")
        case "%":
            rem=input("a%b or b%a: ")
            if rem == "a%b":
                if b != 0:
                    print("The remainder of ",a," divided by ",b," is ",a%b)
                else:
                    print("Error: Division by zero is not allowed.")
            elif rem == "b%a":
                if a != 0:
                    print("The remainder of ",b," divided by ",a," is ",b%a)
                else:
                    print("Error: Division by zero is not allowed.")
            else:
                print("Invalid input")
        case "**":
            exp=input("a**b or b**a: ")
            if exp == "a**b":
                print("The result of ",a," raised to the power of ",b," is ",a**b)
            elif exp == "b**a":
                print("The result of ",b," raised to the power of ",a," is ",b**a)
            else:
                print("Invalid input")
simple_calc()
rep=input("would you like to use the calculator for any other number (yes or no): ")
if rep!="no":
    n=int(input("No. of times you would like to use the calculator: "))
    for i in range(n):
        simple_calc()
print("Thank you for choosing us. Have a great day!!")
```
